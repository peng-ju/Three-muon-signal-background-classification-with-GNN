### Without background

```python
# features required by Yibo
features = ['eventID','truthtrack_px_st3','truthtrack_py_st3','truthtrack_pz_st3', 'truthtrack_x_st3','truthtrack_y_st3','truthtrack_z_st3','truthtrack_px_st1','truthtrack_py_st1','truthtrack_pz_st1', 'truthtrack_x_st1','truthtrack_y_st1','truthtrack_z_st1']

for path in file_path[:]:
    file = uproot.open(path)
    print(path, '\n',file.keys())
    file_key = file.keys()
    
    
    for keys in file_key:
        Event2 = file[keys]
        DATA = pd.DataFrame()
        if features[1] not in Event2.keys():
            break

        for i in range(0,len(features)):
            
            temp = Event2[features[i]].array()
            if 'var' in str(temp.type):
                new_coloumn = [] 
                for j in range(len(temp)):
                    new_coloumn.append(list(temp[j]))
                res = new_coloumn
            else:
                res = np.array(temp)
            
            DATA[features[i]] = res
    
        path.split('.')
        DATA.to_pickle(path[0:-5]+ '_'+ keys[-1] +'.pkl')
```

### Combine signal and background

```python
BG_path = 'DY_simPU_NEW_FULL_Mar14.root'
Sig_path = ['15beam0.3mass_hepmc_displaced.root',
 '15beam0.5mass_hepmc_displaced.root',
 '15beam0.7mass_hepmc_displaced.root',
 '15beam0.9mass_hepmc_displaced.root',
 '20beam0.3mass_hepmc_displaced.root',
 '20beam0.5mass_hepmc_displaced.root',
 '20beam0.7mass_hepmc_displaced.root',
 '20beam0.9mass_hepmc_displaced.root']

def root2pd(path):
    file = uproot.open(path)
    print(path, '\n',file.keys())
    file_key = file.keys()
       
    for keys in [file_key[0]]:
        Event2 = file[keys]
        DATA = pd.DataFrame()
        if features[1] not in Event2.keys():
            break

        for i in range(0,len(features)):
            
            temp = Event2[features[i]].array()
            if 'var' in str(temp.type):
                new_coloumn = [] 
                for j in range(len(temp)):
                    new_coloumn.append(list(temp[j]))
                res = new_coloumn
            else:
                res = np.array(temp)
            
            DATA[features[i]] = res
    
    return DATA

# save background 
df_bg = root2pd(BG_path)
df_bg_lf = df_bg.drop(range(5000), axis=0)
df_bg_lf.to_pickle('DY_simPU_NEW_FULL_Mar14.pkl')

# add background to signal
for path in Sig_path:
    df_sig = root2pd(path)

    for feature in df_sig.columns:
        df_sig[feature] = df_sig[feature] + df_bg[feature]

    path.split('.')
    df_sig.to_pickle(path[0:-5]+ '_'+ keys[-1] +'.pkl')
```
